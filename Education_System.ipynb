```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Expanded_data_with_more_features.csv')
print(df.head())
```

```python
df=df.drop("Unnamed: 0", axis = 1)
df.head()
```

#Gender Distribution

```python
plt.figure(figsize =(5,5))

ax = sns.countplot(data=df, x='Gender', hue='Gender')
ax.bar_label(ax.containers[0])
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
gp=df.groupby('ParentEduc').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
print(gp)
```

```python
plt.figure(figsize =(4,4))
sns.heatmap(gp, annot =True)
plt.show()
```
